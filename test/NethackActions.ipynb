**Codice per testare come funziona l'esecuzioni delle azioni con nle**

* 0 -> Movimento a nord di 1
* 1 -> Movimento a destra di 1
* 2 -> Moviemnto a sud di 1
* 3 -> Movimento a sinistra di 1
* 4 -> Nord est
* 5 -> Sud est
* 6 -> Sud ovest
* 7 -> Nord ovest
* 49 -> Raccoglie oggetto

```python
import sys
sys.path.append('minihack/')
import gym
import nle
from minihack import LevelGenerator
import time
from IPython.display import clear_output


des_file = """
MAZE: "mylevel", ' '
FLAGS:premapped
GEOMETRY:center,center

MAP
|||||||||||||
|...........|
|..|........|
|...........|
|.......|...|
|.....|.....|
|...|....|..|
|||||||||||||
ENDMAP

BRANCH:(1,1,1,1),(0,0,0,0)


GOLD:1,(1,2)
GOLD:1,(7,4)
GOLD:1,(9,4)
GOLD:1,(11,5)
GOLD:1,(6,1)

"""

env = gym.make("MiniHack-Skill-Custom-v0",
    des_file=des_file)
env.reset()
env.render()
```

```python
env.render()
time.sleep(0.5)  # Attendi 500ms

aggent_actions = [2, 49, 4, 1, 1, 1, 1, 49, 5, 2, 2, 49, 4, 5, 49, 5, 1, 49]
for action in aggent_actions:
    env.step(action)
    clear_output(wait=True)
    env.render()
    time.sleep(0.5)  # Attendi 500ms
```
